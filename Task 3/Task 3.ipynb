### **Task: Perform ‘Exploratory Data Analysis’ on dataset ‘SampleSuperstore’**

```python
# Importing all libraries required in this notebook
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
df=pd.read_csv('SampleSuperstore.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize= (10,16))
df.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

```python
plt.figure(figsize= (10,16))
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(5,4))
sns.heatmap(df.corr(), annot=True)
```

```python
plt.figure(figsize=(8,7))
sns.countplot(df['Sub-Category'])
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(8,7))
sns.lineplot(df['Discount'], df['Profit'], data=df)
```

```python
plt.figure(figsize=(8,7))
sns.lineplot(df['Discount'], df['Quantity'], data=df)
```

```python
plt.figure(figsize=(8,7))
sns.lineplot(df['Discount'], df['Profit'], data=df)
```

```python
plt.figure(figsize=(8,7))
sns.lineplot(df['Ship Mode'], df['Profit'], data=df)
```

```python
df.groupby('State')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

```python

plt.figure(figsize= (10,16))
df.groupby('Ship Mode')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('Sale')
plt.show()
```
